```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(1991)
popa = pd.Series(np.random.beta(a=0.9, b=40, size=1000))*100000
popb = pd.Series(np.random.beta(a=0.7, b=40, size=1000))*100000+2000
df = pd.concat([popa,popb],axis=1)
df.columns = ["Population A", "Population B"]
df.plot.kde()
#plt.vlines([popa.mean(), popb.mean()], ymin=0, ymax=0.005, color=["blue","orange"])
```

```python
print(popa.mean())
print(popb.mean())
```

```python
results = []
for i in range(1,101):
    greater = []
    for j in range(1,101):
        if popb.sample(1).values[0] > popa.sample(1).values[0]:
            greater.append(1)
    results.append(sum(greater)/100)
np.mean(results)
```
